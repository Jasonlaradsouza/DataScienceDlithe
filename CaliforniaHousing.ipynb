# Regression Problem

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/instituteofai/ML-101/master/Datasets/cal_housing.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
labels = df[['median_house_value']]
labels.head()
```

```python
features = df[['population','total_rooms']]
features.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(features,labels, test_size=0.15)
```

```python
X_train.shape,X_test.shape
```

```python
Y_train.shape,Y_test.shape
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
predictions = model.predict(X_train)
```

```python
r2 = r2_score(Y_train,predictions)
print("R2 Score :",r2)
```

```python
predictions = model.predict(X_test)
```

```python
r2 = r2_score(Y_test,predictions)
print("R2 Score :",r2)
```

```python
features = df[['longitude',	'latitude',	'housing_median_age',	'total_rooms',	'population',	'households',	'median_income']]
```

```python
df.head()
```

```python
labels = df[['median_house_value']]
labels
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(features,labels, test_size=0.15)
```

```python
X_train.shape,X_test.shape
```

```python
Y_train.shape,Y_test.shape
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import r2_score
predictions = model.predict(X_train)
r2 = r2_score(Y_train,predictions)
print("R2 Score :",r2)
```

```python
predictions = model.predict(X_test)
r2 = r2_score(Y_test,predictions)
print("R2 Score :",r2)
```
