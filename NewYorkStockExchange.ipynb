#New York Stock Exchange

```python
import pandas as pd
```

```python
df=pd.read_csv('fundamentals.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
labels=df[['Total Revenue']]
labels.head()
```

```python
features=df[['Total Assets', 'Total Current Assets', 'Total Current Liabilities', 'Total Equity', 'Total Liabilities', 'Total Liabilities & Equity']]
features.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.15)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

## Training R2 Score

```python
predictions = model.predict(X_train)
```

```python
r2 = r2_score(y_train, predictions)
print("R2 Score :", r2)
```

## Testing R2 Score

```python
predictions = model.predict(X_test)

r2 = r2_score(y_test, predictions)
print("R2 Score :", r2)
```
